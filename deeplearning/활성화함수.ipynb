# 시그모이드

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmaoid(x) : 
    return 1 / (1 + np.exp(-x))
```

```python
x = np.linspace(-10, 10, 100) # -10 부터 10까지 100개 데이터 생성
print(x)
y = sigmaoid(x)

print(sigmaoid(100))

plt.plot(x, y)
plt.show()
```

# 하이퍼볼릭 탄젠트

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def tanh(x) :
    return np.tanh(x)
```

```python
x = np.linspace(-5, 5, 100)
print(x)
y = tanh(x)
print(y)
print(tanh(-5))
plt.plot(x, y)
plt.show()
```

# ReLU

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def relu(x) : 
    return np.maximum(0, x)

x = np.linspace(-5, 5, 1000)
y = relu(x)

plt.plot(x, y)
plt.show()
```

# Leaky ReLU

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def leaky_relu(x, alpha=0.01) : 
    return np.where(x >= 0, x, x * alpha)

x_temp = np.linspace(-5, 5, 100)
y_tmep = leaky_relu(x_temp)

plt.plot(x_temp,y_tmep)
plt.show()
```

# 소프트맥스

```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def softmax(z) : 
    return np.exp(z) / np.sum(np.exp(z))

z = np.linspace(-10, 10, 100) # -10 부터 10까지 100개의 값을 생성하라 !!
y = softmax(z)

plt.plot(z, y)
plt.show()
```
