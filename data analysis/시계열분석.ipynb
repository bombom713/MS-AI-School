## 추세 분석, 계절성 분석, 자기 상관성 분석

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_acf
```

```python
# 시계열 데이터 생성 
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=1000, freq='D')
values = np.random.randn(1000)
```

```python
# 데이터프레임 생성
data = pd.DataFrame({'Date' : dates, 'Value' : values})
```

```python
# 데이터프레임을 csv 파일로 저장
data.to_csv('./sample_time_data.csv', index=False)
```

```python
# 저장한 csv 파일 읽기
data = pd.read_csv('./sample_time_data.csv', parse_dates=['Date'], index_col='Date')
```

```python
# 데이터 확인
print(data)
```

```python
# 시계열 데이터 시각화
plt.plot(data.index, data['Value'])
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Data')
plt.show()
```

```python
# 추세 분석
result = seasonal_decompose(data['Value'], model='additive', period=365)
result.plot()
plt.show()
```

```python
# 계절성 분석
seasonal_component = result.seasonal
plt.plot(seasonal_component)
plt.xlabel('Date')
plt.ylabel('Seasonal Component')
plt.title('Seasonal Analysis')
plt.show()
```

```python
# 자기 상관성 분석
plot_acf(data['Value'], lags=50)
plt.xlabel('Lag')
plt.ylabel('Autocorrelation')
plt.title('Autocorrelation Function')
plt.show()
```

```python
# 시계열 데이터 시각화 기법: 시계열 분해
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
```

```python
# 시계열 데이터 생성
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.randn(100)
```

```python
# 데이터프레임 생성
data = pd.DataFrame({'Date' : dates, 'Value' : values})
```

```python
# 시계열 분해
result = seasonal_decompose(data['Value'], model='additive', period=7)
```

```python
# 분해 결과 시각화
plt.figure(figsize=(10, 8))
```

```python
# 원본 데이터
plt.subplot(4, 1, 1)
plt.plot(data['Date'], data['Value'])
plt.title('Original Data')
plt.ylabel('Value')
```

```python
# 추세(Trend)
plt.subplot(4, 1, 2)
plt.plot(data['Date'], result.trend)
plt.title('Trend')
plt.ylabel('Trend')
```

```python
# 계절성(Seasonality)
plt.subplot(4, 1, 3)
plt.plot(data['Date'], result.seasonal)
plt.title('Seasonality')
plt.ylabel('Seasonality')
```

```python
# 불규칙성(Irregularity)
plt.subplot(4, 1, 4)
plt.plot(data['Date'], result.resid)
plt.title('Irregularity')
plt.ylabel('Irregularity')

plt.tight_layout()
plt.show()
```

```python
# 결측치 전처리: 스플라인 보간법
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
```

```python
# 시계열 데이터 생성
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=1000, freq='D')
values = np.random.randn(1000)
```

```python
# 결측치 생성
data.loc[100 : 200, 'Value'] = np.nan
```

```python
# 결측치 생성 확인 체크
print(data.loc[100 : 200])
```

```python
# 스플라인 보간법은 유효한(null 값이 아닌) 데이터에 대해서만 적용할 수 있기 때문에, 결측치를 제거하거나 대체하는 작업이 필요합니다.
# 보간법을 적용하여 결측치를 처리해봅시다.
data['Value_filled'] = data['Value'].interpolate(method='linear')
print(data.loc[100 : 200])
```

```python
# 스플라인 보간법 적용
sql = CubicSpline(data.index, data['Value_filled'])
data['Value_spline'] = sql(data.index)
print(data.loc[100 : 200])
```

```python
# 시계열 데이터 시각화
plt.plot(data['Date'], data['Value'], 'o-', label='Original')
plt.plot(data['Date'], data['Value_spline'], 'x-', label='Spline Interpolation')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Spline Interpolation')
plt.legend()
plt.show()
```

## 시계열 데이터 전처리: 이상치

이동평균과 이동표준편차

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', 500)
```

```python
# 시계열 데이터 생성
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=500, freq='D')
values = np.random.randn(500)
```

```python
# 데이터 프레임 생성
data = pd.DataFrame({'Date' : dates, 'Value' : values})
```

```python
# 이동 평균과 이동 표준편차 계산
window_size = 7  # 7일(일주일) 주기로 계산
data['MovingAverage'] = data['Value'].rolling(window=window_size).mean()
data['MovingStd'] = data['Value'].rolling(window=window_size).std()
print(data['MovingAverage'])
print(data['MovingStd'])
```

```python
# 임계값 설정
threshold_multiplier = 2  # 임계값 배수 (이상치 범위 설정)
data['Threshold'] = data['MovingAverage'] + threshold_multiplier * data['MovingStd']
print(data['Threshold'])
```

```python
# 이상치 탐지
data['Outlier'] = data['Value'] > data['Threshold']  # 이상치일 경우 True 반환
print(data['Outlier'])
```

지수평활법

* 잔차: 관찰값과 예측값의 차이
* 차분: 관찰값의 먼저값 - 관찰값의 나중값의 차이

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.holtwinters import ExponentialSmoothing
```

```python
# 시계열 데이터 생성
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = [10, 12, 11, 15, 9, 13, 8, 14, 10, 16] + [10] * 90  # 이상치가 포함된 데이터
```

```python
# 데이터프레임 생성
data = pd.DataFrame({'Date' : dates, 'Value' : values})
```

```python
# 지수 평활법 모델 학습
model = ExponentialSmoothing(data['Value'], trend='add')  # trend = 'add'의 다른 방식으로는 'mul'(곱셈), 'additive'(가법), 'multiplicative'(승법)이 있음
model_fit = model.fit()
```

```python
# 추세 계산
trend = model_fit.predict(start=0, end=len(data['Value'])-1)  # end값은 데이터의 전체 기간을 계산하는 방법
```

```python
#추세에서의 편차 계산
deviation = data['Value'] - trend
```

```python
# 이상치 탐지
threshold = 2  # 이상치 임계값 설정
outliers = abs(deviation) > threshold
```

```python
# 시계열 데이터 시각화
plt.plot(data['Date'], data['Value'], label='Original')
plt.plot(data['Date'], trend, label='Trend')
plt.scatter(data[outliers]['Date'], data[outliers]['Value'], color='red', label='Outliers')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Outlier Detection with Exponential Smoothing')
plt.legend()
plt.show()
```

Z-점수

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
# 시계열 데이터 생성
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.randn(100) * 10  # 정규분포 데이터 생성
```

```python
# 이상치 추가
values[20] = 50
```

```python
# 데이터 프레임 생성
data = pd.DataFrame({'Date' : dates, 'Value' : values})
```

```python
# 이상치 탐지
z_scores = np.abs(stats.zscore(data['Value']))  # Z-score 계산
threshold = 3  # 이상치 임계값 설정
outliers = z_scores > threshold  # Z-score가 임계값을 초과하는지 확인
```

```python
# 이상치 시각화
plt.plot(data['Date'], data['Value'], label='Original')
plt.scatter(data[outliers]['Date'], data[outliers]['Value'], color='red', label='Outliers')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Outlier Detection with Z-Score in Time Series Data')
plt.legend()
plt.show()
```

```python
# 이상치 출력
outlier_values = data[outliers]['Value']
print("Detected Outliers:")
print(outlier_values)
```

##  시계열 특성을 고려한 이상치 탐지

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
```

```python
# 시계열 데이터 생성
np.random.seed(42)
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.randn(100) * 10  # 정규분포 데이터 생성
```

```python
# 이상치 추가
values[20] = 50
```

```python
# 데이터 프레임 생성
data = pd.DataFrame({'Date' : dates, 'Value' : values})
print("이상치 데이터 추가 확인 체크 >>", data['Value'].iloc[20])
```

```python
# 이상치 탐지
z_scores = np.abs(stats.zscore(data['Value']))  # z-score 계산
threshold = 3  # 이상치 임계값 설정
outliers = z_scores > threshold  # z-score가 임계값을 초과하는지 확인
```

```python
# 이상치 처리
data.loc[outliers, 'Value'] = np.nan  # 이상치를 결측치로 처리
```

```python
# 결측치 보간
data['Value'] = data['Value'].interpolate()
```

```python
# 시계열 데이터 시각화
plt.plot(data['Date'], data['Value'], label='Original')
plt.scatter(data[outliers]['Date'], data[outliers]['Value'], color='red', label='Outliers')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Outlier Detection and Handling in Time Series Data')
plt.legend()
plt.show()
```

```python
# 처리된 데이터 확인
print("이상치 처리된 데이터 확인 >>", data['Value'].iloc[20])
```
