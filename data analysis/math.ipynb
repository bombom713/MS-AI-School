```python
# 신경망 기초
# 파이썬으로 딥러닝의 신경망 공식을 배우고 시각화 해봅시다.
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1차함수부터 봅시다. y = ax + b: a는 기울기 b는 y의 절편
def linear_function(x):
    a=0.5
    b=2
    
    return a*x+b
```

```python
print(linear_function(5))
```

```python
x=np.arange(-5,5,0.1)
y=linear_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
```

```python
# 이차합수를 배워봅시다. y = ax^2 + bx + c
def quadratic_function(x):
    a=1
    b=-1
    c=-2
    
    return a*x**2+b*x+c
```

```python
print(quadratic_function(2))
```

```python
x=np.arange(-5,5,0.1)
y=quadratic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
```

```python
# 삼차함수(다항함수)를 배워봅시다. y = ax^3 + bx^2 + cx + d
def cubic_function(x):
    a=4
    b=0
    c=-1
    d=-8
    
    return a*x**3+b*x**2+c*x+d
```

```python
x=np.arange(-5,5,0.1)
y=cubic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Function')
```

```python
# 논리게이트를 만들어봅시다. 제일 먼저 AND 게이트를 만들어봅시다.
def AND(a,b):
    input=np.array([a,b])
    weights=np.array([0.4,0.4])
    bias=-0.6
    value=np.sum(input*weights)+bias
    
    if value<=0:
        return 0
    else:
        return 1
```

```python
print(AND(0,0))
print(AND(0,1))
print(AND(1,0))
print(AND(1,1))
```

```python
# OR 게이트를 만들어봅시다.
def OR(a,b):
    input=np.array([a,b])
    weights=np.array([0.4,0.4])
    bias=-0.3
    value=np.sum(input*weights)+bias
    
    if value<=0:
        return 0
    else:
        return 1
```

```python
print(OR(0,0))
print(OR(0,1))
print(OR(1,0))
print(OR(1,1))
```

```python
# NAND 게이트를 만들어봅시다.
def NAND(a,b):
    input=np.array([a,b])
    weights=np.array([-0.5,-0.5])
    bias=0.7
    value=np.sum(input*weights)+bias
    
    if value<=0:
        return 0
    else:
        return 1
```

```python
print(NAND(0,0))
print(NAND(0,1))
print(NAND(1,0))
print(NAND(1,1))
```

```python
# Exclusive OR 게이트를 만들어봅시다.
def XOR(x1,x2):
    s1=NAND(x1,x2)
    s2=OR(x1,x2)
    y=AND(s1,s2)
    
    return y
```

```python
print(XOR(0,0))
print(XOR(0,1))
print(XOR(1,0))
print(XOR(1,1))
```

```python
# Activation Function(활성화함수)를 배워봅시다. 활성화함수는 입력값을 각 문제에 알맞은 출력값을 반환하기 위해서 필요하며 각 레이어마다 지정해줘야 합니다.
# 맨 처음으로 Step Function(계단함수)를 배워봅시다. 이 함수는 0부터 0.999...까지는 0을 반환하다가 비로소 1을 넘어야 1을 반환하는 함수입니다.
def step_function(x):
    if x>0:
        return 1
    else:
        return 0
```

```python
print(step_function(0))
print(step_function(-1))
print(step_function(0.1))
```

```python
def step_function_for_numpy(x):
    y = x>0
    return y.astype(np.int)
```

```python
a=np.array([5,3,-4,2,0])
print(step_function_for_numpy(a))
```

```python
# Sigmoid Function(시그모이드 함수)를 배워봅시다. 시그모이드는 개냐 고양이냐 등과 같이 2진으로 나눠지는 문제에서 쓰이며 확실한건 먼저 분류해주고 애매한 건 가운데 기준선을 그어 나눠주는 함수입니다.
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
print(sigmoid(3))
print(sigmoid(10))
print(sigmoid(100))
print(sigmoid(-3))
print(sigmoid(-100))
```

```python
# Step Function과 Sigmoid Function을 비교해보는 시각화를 해봅시다.
x = np.arange(-5,5,0.01)
y1 = sigmoid(x)
y2 = step_function_for_numpy(x)

plt.plot(x, y1, 'r-', x, y2, 'b--')
```

```python
# ReLU함수를 배워봅시다. ReLU함수는 가장 많이 쓰이는 함수입니다. ReLU함수는 마이너스 값이 필요없을 때 사용합니다. 마이너스 값을 전부 0으로 반환합니다.
def ReLU(x):
    if x>0:
        return x
    else:
        return 0
```

```python
print(ReLU(5))
print(ReLU(-5))
```

```python
# Identity Function(항등함수)는 들어오는 값을 그대로 반환하는 함수기 때문에 사실 함수가 없는 것이나 다름 없습니다. 주로 회귀문제에서 사용합니다. 회귀에서는 값이 들어와서 왜곡되면 안되기 때문입니다.
def identity_function(x):
    return x
```

```python
# Softmax(소프트맥스)함수는 출력의 결과를 다 합쳐서 1로 만들어 반환합니다. 입력값이 아무리 커도 출력값은 다 합치면 1이됩니다. 반환값을 읽는 것은 4일 확률이 80%(0.8)입니다. 이렇게 읽으면 됩니다. 소프트맥스는 큰 값은 더 크게 작은 값은 더 작게 해석해서 반환합니다.
# 다중문제는 소프트맥스, 싱글문제(2진문제)는 시그모이드, 일반적으론 렐루, 리그리션은 아이덴티티 함수를 사용합니다.
def softmax(x):
  exp_x = np.exp(x)
  sum_exp_x = np.sum(exp_x)
  y = exp_x / sum_exp_x

  return y
```

```python
x = np.array([0.3, 0.2, 3.0, -1, 2])
print(softmax(x))
print(np.sum(softmax(x)))
```
