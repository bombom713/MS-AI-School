```python
import numpy as np
import matplotlib.pyplot as plt

"""
계산방법
각 숫자 빈도수 체크
예를 들어 5라는 숫자의 빈도수는 4번
4/11(데이터 총 수) = 0.36363636
"""

# 예시 데이터 생성
data = np.array([1, 2, 3, 3, 4, 4, 4, 5, 5, 5, 5])

# 데이터에서 각 값의 빈도수 계산
value_counts = np.bincount(data)  # bincount는 0부터 세기 때문에 0이 array에 없어도 0부터 5까지의 빈도수를 세서 0의 빈도수도 출력됌
print(value_counts)

# 각 값의 상대빈도수 계싼
relative_frequencies = value_counts / len(data)
print(len(data))
print(relative_frequencies)
```

```python
# bincount 모듈을 쓰면 튜플형태로 값을 반환하고 각 값이 얼마의 빈도수를 갖는지는 알아볼 수 없음
# 리스트 값마다 얼마의 빈도수를 갖는지 알아보려면 아래처럼 함수를 만들어서 출력하면 됌

from collections import Counter

def print_frequency(numbers):
    # 빈도수 계산
    counts = Counter(numbers)
    
    # (빈도수, 수) 쌍의 리스트 생성
    frequency_list = [(count, number) for number, count in counts.items()]
    
    # 빈도수를 기준으로 정렬
    sorted_list = sorted(frequency_list)
    
    # 결과 출력
    for count, number in sorted_list:
        print(f"수: {number}, 빈도수: {count}")

# 예시 입력 데이터
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4]

# 함수 호출
print_frequency(data)
```

```python
# 막대 그래프로 상대빈도수 시각화
plt.bar(np.arange(len(value_counts)), relative_frequencies)
plt.xticks(np.arange(len(value_counts)), np.arange(len(value_counts))+1)
plt.xlabel('Value')
plt.ylabel('Relative Frequency')
plt.title('Relative Frequency Bar Chart')
plt.show()
```

```python
# bincount를 사용함으로써 0이 출력되는 것을 막으려면 unique 함수를 써야 됌
# 예시 데이터 생성
data = np.array([1, 2, 3, 3, 4, 4, 4, 5, 5, 5, 5])

# 데이터에서 각 값의 빈도수 계산
index, value_counts = np.unique(data, return_counts=True)
print(value_counts)

# 각 값의 상대빈도수 계싼
relative_frequencies = value_counts / len(data)
print(len(data))
print(relative_frequencies)
```

```python
# 막대 그래프로 상대빈도수 시각화
# x축의 범위를 array의 갯수로 설정하면 총 5개이지만 인덱스번호로는 0~4로 표기되기 때문에, 각각의 라벨을 정해주고 출력해야 0이 그래프에서 나오지 않음
X = np.arange(len(value_counts))
xlabel = ['1','2','3','4','5']

plt.bar(np.arange(len(value_counts)), relative_frequencies)
plt.xticks(X, xlabel)
plt.xlabel('Value')
plt.ylabel('Relative Frequency')
plt.title('Relative Frequency Bar Chart')
plt.show()
```
