```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# 예제 데이터 생성
X = np.random.rand(100, 3)  # 100개의 샘플과 3개의 변수로 구성된 데이터
```

```python
# PCA 모델 생성 및 적합
pca = PCA(n_components=2)  # 주성분을 2개로 설정
pca.fit(X)
```

```python
# 변환된 데이터
transformed_data = pca.transform(X)
```

```python
# 주성분의 설명력 확인
explained_variance_ratio = pca.explained_variance_ratio_
print("주성분의 설명력:", explained_variance_ratio)
```

```python
# 주성분의 가중치 확인
components = pca.components_
print("주성분의 가중치:\n", components)
```

```python
# 변환된 데이터 시각화
plt.scatter(transformed_data[:, 0], transformed_data[:, 1])
plt.title('PCA result')
plt.show()
```

#### PCA를 활용한 차원 축소 및 데이터 시각화

```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# 원본 데이터 생성
data = np.random.rand(100, 3)  # 100개의 샘플과 3개의 변수로 구성된 데이터
```

```python
# PCA 모델 생성 및 학습
pca = PCA(n_components=2)  # 축소할 차원 수를 2로 설정
reduced_data = pca.fit_transform(data)
```

```python
# 축소된 데이터 시각화
plt.scatter(reduced_data[:, 0], reduced_data[:, 1])
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA - Dimensionality Reduction')
plt.show()
```
