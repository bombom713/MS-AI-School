```python
import pandas as pd
import matplotlib.pyplot as plt

# 데이터셋 불러오기
data = pd.read_csv('C:\Data Analysis\BlackFriday.csv')

# User_ID별 구매 총 금액 계산
user_total_spent = data.groupby('User_ID')['Purchase'].sum()
```

```python
plt.hist(user_total_spent, bins=50)  # bins는 구간의 갯수. 50개로 나눔.
plt.title('Histogram of Total Purchase Amount by User')
plt.xlabel('Total Purchase Amount')
plt.ylabel('Frequency')
plt.show()
```

```python
# 밀도 그래프 그리기
plt.figure()
user_total_spent.plot.kde()
plt.title('Density Plot of Total Purchase Amount by User')
plt.xlabel('Total Purchase Amount')
plt.show()
```
