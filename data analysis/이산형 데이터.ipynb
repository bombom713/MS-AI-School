```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('C:\Data Analysis\college_data.csv')

# top10perc 컬럼은 대학교의 학생 중 상위 ㅣ10% 학생의 비율을 나타내는 이산형 데이터입니다.
# 이 컬럼을 사용하여 이산형 데이터 시각화를 실습해보겠습니다.
print(data['top10perc'].unique())  # unique는 index를 없애고 값만 나열해주는 명령어
```

```python
# top10perc 막대 그래프 그리기
plt.figure(figsize=(10,6))
plt.bar(data['private'], data['top10perc'])
plt.title('Bar Chart of Top 10% Scores by Private/Public')
plt.xlabel('Private/Public')
plt.ylabel('Top 10% Scores')
plt.show()
```

```python
# top10perc 원 그래프 그리기
plt.figure(figsize=(10,6))
data.groupby('private')['top10perc'].sum().plot(kind='pie', autopct='%1.1f%%')
plt.title('Pie Chart of Top 10% Scores by Private/Public')
plt.ylabel('')
plt.show()
```

```python
# top10perc 히스토그램 그리기
plt.figure(figsize=(10,6))
plt.hist(data['top10perc'], bins=20)
plt.title('Histogram of Top 10% Scores')
plt.xlabel('Top 10% Scores')
plt.ylabel('Frequency')
plt.show()
```
