```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
```

```python
# 이항분포 모델링
n = 10  # 시행횟수
p = 0.5  # 성공확률

# 이항분포 시뮬레이션
num_simulations = 1000  # 시뮬레이션 횟수
successes = np.random.binomial(n, p, num_simulations)

# 이항분포 시뮬레이션 결과 시각화
plt.hist(successes, bins=11, range=(0, 10), alpha=0.8)
plt.show()
```
